Create a Python program that converts between different units of measurement.



```python
def convert(option, num):
    """
    Converts units based on the selected option.

    Parameters:
    option (int): conversion option selected by the user
    num (float): value to convert

    Returns:
    float: converted value
    """

    if option == 1:
        return num * 3.28
    elif option == 2:
        return num / 3.28
    elif option == 3:
        return num * 2.2
    elif option == 4:
        return num / 2.2
    elif option == 5:
        return num * 0.26
    elif option == 6:
        return num / 0.26
    else:
        return None


print("Unit Converter")

print("1. Meter to Feet")
print("2. Feet to Meter")
print("3. Kg to Pounds")
print("4. Pounds to Kg")
print("5. Liter to Gallon")
print("6. Gallon to Liter")

try:
    option = int(input("Enter option: "))
    num = float(input("Enter value: "))

    ans = convert(option, num)

    if ans is None:
        print("Unsupported conversion option.")
    else:
        print("Converted value is:", ans)

except ValueError:
    print("Invalid input. Please enter numeric values.")
```

Create a Python program that performs various mathematical operations on a list of numbers.

```python

def find_sum_list(nums):
    """
    calculate sum of numbers in list
    nums: list of numbers
    return: total sum
    """
    total = 0
    for n in nums:
        total = total + n
    return total


def find_avg_list(nums):
    """
    calculate average of numbers in list
    nums: list of numbers
    return: average value
    """
    total = find_sum_list(nums)
    return total / len(nums)


def find_max_list(nums):
    """
    find largest number in list
    nums: list of numbers
    return: maximum number
    """
    m = nums[0]
    for n in nums:
        if n > m:
            m = n
    return m


def find_min_list(nums):
    """
    find smallest number in list
    nums: list of numbers
    return: minimum number
    """
    m = nums[0]
    for n in nums:
        if n < m:
            m = n
    return m


print("Simple list calculator")
print("opt1 find sum ")
print("opt2 find average")
print("opt3 find max")
print("opt4 find min")

try:
    op = int(input("choose operation: "))
    text = input("type numbers separated by space: ")

    parts = text.split()
    nums = []

    for p in parts:
        nums.append(float(p))

    if len(nums) == 0:
        print("you did not enter any numbers")

    elif op == 1:
        print("sum =", find_sum_list(nums))

    elif op == 2:
        print("average =", find_avg_list(nums))

    elif op == 3:
        print("max number =", find_max_list(nums))

    elif op == 4:
        print("min number =", find_min_list(nums))

    else:
        print("invalid operation")



except ValueError:
    print("please enter numbers only")
```

Exercise on List Manipulation:

```python
# get every other element
def extract_every_other(lst):
    return lst[::2]

# get sublist from start to end
def get_sublist(lst, start, end):
    return lst[start:end+1]

# reverse the list
def reverse_list(lst):
    return lst[::-1]

# remove first and last items
def remove_first_last(lst):
    return lst[1:-1]

# get first n items
def get_first_n(lst, n):
    return lst[:n]

# get last n items
def get_last_n(lst, n):
    return lst[-n:]

# reverse skipping every second element
def reverse_skip(lst):
    return lst[-2::-2]


# testing with my own random list
my_list = [11, 7, 23, 5, 19, 14]
print("Every other element :", extract_every_other(my_list))
print("Sublist from index 2 to 4 :", get_sublist(my_list, 2, 4))
print("Reversed list :", reverse_list(my_list))
print("Without first & last :", remove_first_last(my_list))
print("First 3 items :", get_first_n(my_list, 3))
print("Last 2 items :", get_last_n(my_list, 2))
print("Reverse skipping every second :", reverse_skip(my_list))
```

Exercise on Nested List

```python

def flatten(lst):
    result = []
    for sub in lst:
        for x in sub:
            result.append(x)
    return result

def access_nested_element(lst, idxs):
    temp = lst
    for i in idxs:
        temp = temp[i]
    return temp

def sum_nested(lst):
    total = 0
    for x in lst:
        if isinstance(x,list):
            total += sum_nested(x)
        else:
            total += x
    return total

def remove_element(lst, elem):
    new_lst = []
    for sub in lst:
        new_sub = []
        for x in sub:
            if x != elem:
                new_sub.append(x)
        new_lst.append(new_sub)
    return new_lst

# find max element
def find_max(lst):
    max_val = float('-inf')
    for x in lst:
        if isinstance(x,list):
            val = find_max(x)
        else:
            val = x
        if val > max_val:
            max_val = val
    return max_val

# count occurrences of elem
def count_occurrences(lst, elem):
    count = 0
    for x in lst:
        if isinstance(x,list):
            count += count_occurrences(x, elem)
        else:
            if x == elem:
                count += 1
    return count

# deep flatten any depth
def deep_flatten(lst):
    flat = []
    for x in lst:
        if isinstance(x,list):
            flat += deep_flatten(x)
        else:
            flat.append(x)
    return flat

# average of all elements
def average_nested(lst):
    flat = deep_flatten(lst)
    if flat:
        return sum(flat)/len(flat)
    else:
        return 0


nested1 = [[7,13],[22,5],[19]]
nested2 = [[8,17],[3,[14,2]],11]
nested3 = [[9,4],[2,18],[6,12]]

print("Flatten:", flatten(nested1))
print("Element [1,1]:", access_nested_element([[10,11,12],[20,21,22],[30,31,32]], [1,1]))
print("Sum:", sum_nested(nested2))
print("Remove 2:", remove_element([[9,2],[2,18],[6,2]], 2))
print("Max:", find_max(nested2))
print("Count 2:", count_occurrences(nested3, 2))
print("Deep flatten:", deep_flatten([[[3,7],[1,9]],[[5,2],[8,6]]]))
print("Average:", average_nested([[4,15],[8,12],[20,7]]))
```
